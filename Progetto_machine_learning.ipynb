```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
import itertools
import os
from google.colab.patches import cv2_imshow
```

```python
#loading custom weigths just trained
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/yolov5/runs/train/exp3/weights/best.pt', force_reload=True)
```

```python
%matplotlib inline
plt.close('all')
fig, ax = plt.subplots(1,3, figsize=(20,10))
imgs = os.listdir('/content/yolov5/Road-signs-11/test/images')

for idx in range(3):
    imgname = np.random.choice(imgs)
    img = cv2.imread(f'/content/yolov5/Road-signs-11/test/images/{imgname}')
    results = model(img)
    ax[idx].imshow(cv2.cvtColor(np.squeeze(results.render()), cv2.COLOR_BGR2RGB))
```
